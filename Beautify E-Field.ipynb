```python
import pandas as pd
```

```python
main_path = "/mnt/d/ElectricField/"
r_min = -6 
r_max = 6
z_min = -1
z_max = 7
sep = 0.0025

def beautify(filename,param):
    
    voltage = param[0]
    direction = param[1]
    
    df = pd.read_csv(main_path + filename,skiprows=13,header=None)
    df = df.fillna(0)
    
    df[0][0] = r_min
    df[1][0] = r_max
    df[2][0] = (r_max-r_min)/sep + 1
    
    df[0][1] = z_min
    df[1][1] = z_max
    df[2][1] = (z_max-z_min)/sep + 1
    
    df.to_csv(main_path + "sorted/" + direction + "_Efield_abalone_"+str(voltage)+".csv",header=None,index = None)
    print(df)
```

```python
beautify("Electric_x_225.csv",[225,"x"])
beautify("Electric_y_225.csv",[225,"z"])
beautify("Electric_x_20.csv",[20,"x"])
beautify("Electric_y_20.csv",[20,"z"])
```
